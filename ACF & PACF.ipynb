```python
import pandas as pd 
from statsmodels.tsa.stattools import acf,pacf 
from statsmodels.graphics.tsaplots import plot_acf , plot_pacf
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=250
sns.set_style('darkgrid')
```

```python
#Plot the data 
df=pd.read_csv(r'https://raw.githubusercontent.com/aishwaryamate/Datasets/main/AirPassengers.csv',index_col=0,parse_dates=True)
df
```

```python
df.info()
```

```python
plt.plot(df['Passengers'])
```

```python
#First-order differences
df['First-Order']=df['Passengers'].diff(periods=1)
df
```

```python
#Drop na 
df.dropna(inplace=True)
```

```python
df
```

```python
#Plot
plt.plot(df['Passengers'],color='red',label='Passenger')
plt.plot(df['First-Order'],color='green',label='First-Order')
plt.legend()
```

```python
# calculate Autocorrelation 
```

```python
acf_values=acf(df['First-Order'])
acf_values
```

```python
#Plot ACF
plot_acf(df['First-Order'],lags=30);
```

```python
#Calculate PACF
pacf_val=pacf(df['First-Order'])
pacf_val
```

```python
#Plot PACF
plot_pacf(df['First-Order'],lags=30);
```

# Interpretation of ACF anf PACF plots

. If the ACF plot declines gradually and the PACF drops instantly, use Auto Regressive model.

. If the ACF plot drops instantly and the PACF declines gradually, use Moving Average model.

. If both ACF and PACF decline gradually, combine Auto Regressive and Moving Average models (ARMA).

. If both ACF and PACF drop instantly (no significant lags), itâ€™s likely you wonâ€™t be able to model the time series.
